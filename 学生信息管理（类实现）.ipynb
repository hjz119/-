```python
# -*- coding:utf-8 -*-

import os

#定义学生类
class Student:
    #类似java的构造器
    def __init__(self,id,name,age,sex,score,grade):
        self.id = id
        self.name = name
        self.age = age
        self.sex=sex
        self.score=score
        self.grade=grade

    #相当于java的toString()方法
    def __str__(self):
        #msg = "{'id':" + "'"+self.id +"'"+ ",'name':" + "'" +self.name + "'" + ",'age':" + "'" + self.age+"'}"
        msg = "学生信息：id=" + self.id + ",name=" + self.name + ",age=" + self.age + ",sex=" + self.sex + ",score=" + self.score + ",grade=" + self.grade
        return msg

    #获取id
    def getId(self):
        return self.id
    #获取name
    def getName(self):
        return self.name
    #获取age
    def getAge(self):
        return self.age
    #获取sex
    def getSex(self):
        return self.sex
    #获取score
    def getScore(self):
        return self.score
    #获取grade
    def getGrade(self):
        return self.grade
    #设置name
    def setName(self,name):
        self.name = name
    #设置age
    def setAge(self,age):
        self.age = age
     #设置sex
    def setSex(self,sex):
        self.sex = sex
     #设置score
    def setScore(self,score):
        self.score = score
    #设置grade
    def setGrade(self,grade):
        self.grade = grade
# 添加学生信息
def addStu(array):
    "添加学生信息"
    id = input("请输入学生学号：")
    for i in range(len(array)):
        stu2 = array[i]
        if id == stu2.getId():
            print("该学号已存在，不能重复添加")
            return
    name = input("请输入学生姓名：")
    age = input("请输入学生年龄：")
    sex = input("请输入学生性别：")
    score = input("请输入学生成绩：")
    grade = input("请输入学生年级：")
    stu = Student(id,name,age,sex,score,grade)
    array.append(stu)  # 把单个学生添加到总列表中
    print("添加成功:",stu)

# 删除学生信息
def delStu(array):
    "删除学生信息"
    id = input("请输入要删除的学生学号：")
    for i in range(len(array)):
        stu2 = array[i]
        if id == stu2.getId():
            del array[i]
            return 0
    return 1

# 修改学生信息
def updateStu(array):
    "修改学生信息"
    id = input("请输入要修改的学生学号：")
    for i in range(len(array)):
        stu2 = array[i]
        if id == stu2.getId():
            name = input("请输入要修改的学生姓名：")
            age = input("请输入要修改的学生年龄：")
            sex = input("请输入要修改的学生性别：")
            score = input("请输入要修改的学生成绩：")
            grade = input("请输入要修改的学生年级：")
            stu2.setName(name)
            stu2.setAge(age)
            stu2.setSex(sex)
            stu2.setScore(score)
            stu2.setGrade(grade)
            print("修改成功")
            return
    print("找不到该学号，没法修改")

def sortStu(array):
    "排序学生信息"
    key1 = input("请输入排序关键字：")
    if key1 == "学号":
        array.sort(key=lambda x:x.getId())
    elif key1 == "姓名":
        array.sort(key=lambda x:x.getName())
    elif key1 == "年龄":
        array.sort(key=lambda x:x.getAge())
    elif key1 == "性别":
        array.sort(key=lambda x:x.getSex())
    elif key1 == "成绩":
        array.sort(key=lambda x:x.getScore())
    elif key1 == "年级":
        array.sort(key=lambda x:x.getGrade())
    else:
        print("关键字不存在，排序失败")
    printStuInfo(array)
    
# 查询学生信息
def selectStu(array):
    "查询学生信息"
    id = input("请输入要查询的学生学号：")
    for i in range(len(array)):
        stu2 = array[i]
        if id == stu2.getId():
            print("查询到的学生信息：",stu2)
            return
    print("查询失败，查不到该学生信息")
    return
#打印学生信息
def printStuInfo(array):
    for i in range(len(array)):
        stu = array[i]
        print(stu)

print("=="*30)
print("欢迎使用学生管理系统")
print("1.添加学生信息")
print("2.删除学生信息")
print("3.修改学生信息")
print("4.查询学生信息")
print("5.排序学生信息")
print("6.退出系统")
print("=="*30)
flag = 0
array = [] #定义list用于保存学生信息
filename = 'write_data.txt' #文件名
if not os.path.exists(filename) : # 判断文件是否存在
    file = open(filename, 'w') # 不存在就创建文件
    file.close()
f = open(filename, "r")
content = f.readlines()
print("文件内容：",content)
array.extend(content)
array_temp = [] # 临时变量
for i in range(len(array)): # 遍历转成学生对象
    print("第"+str(i)+"行:", array[i])
    if isinstance(array[i], str):  # 判断是否为字符串
        strArray = str(array[i]).split(",")
        id = strArray[0]
        name = strArray[1]
        age = strArray[2]
        sex = strArray[3]
        score = strArray[4]
        grade = strArray[5].replace("\n","")
        student = Student(id,name,age,sex,score,grade)#创建学生对象
        array_temp.append(student)
del array
array = array_temp
while flag != 1:
    step = input("请输入你的操作：")
    step = int(step)
    if step == 1:
        addStu(array)
        #print("学生信息打印：", array)
    elif step == 2:
        num = delStu(array)
        if num == 0:
            print("删除成功")
        elif num == 1 or num == 2:
            print("删除失败")
        printStuInfo(array)
    elif step == 3:
        updateStu(array)
        printStuInfo(array)
    elif step == 4:
        selectStu(array)
    elif step == 5:
        sortStu(array)
    elif step == 6:
        flag = 1
        with open(filename, 'w') as f:  # 如果filename不存在会自动创建， 'w'表示写数据，写之前会清空文件中的原有数据！
            for i in range(len(array)):
                if i == len(array)-1 :
                    stu =array[i]
                    f.write(stu.getId() + ","+stu.getName()+","+stu.getAge()+","+stu.getSex()+","+stu.getScore()+","+stu.getGrade())
                else:
                    stu = array[i]
                    f.write(stu.getId() + ","+stu.getName()+","+stu.getAge()+","+stu.getSex()+","+stu.getScore()+","+stu.getGrade()+"\n")
            f.close()
    else:
        print("输入指令错误，请重新输入！！")
print("退出系统成功")
```
