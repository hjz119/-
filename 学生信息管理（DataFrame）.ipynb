```python
import pandas as pd
import xlrd
filename='test.xlsm'
data=pd.read_excel(filename,index_col=u'学号')
data
print("=="*30)
print("欢迎使用学生管理系统")
print("1.添加学生信息")
print("2.删除学生信息")
print("3.修改学生信息")
print("4.查询学生信息")
print("5.排序学生信息")
print("6.退出系统")
print("=="*30)
flag=0
#data=data.append(pd.DataFrame({'学号':12,'姓名':'李华','年龄':18,'性别':'男','成绩':60,'年级':2019}))
#
# print(data.index)
# print(12 not in data.index)
# print(data)
while flag != 1:

    step = input("请输入你的操作：")
    try:
        step = int(step)
    except BaseException:
        print("发生异常，输入的不是数字类型")
        break
    if step == 1:
        try:
            id = int(input("请输入增加学生学号："))
            if id in data.index:
                print("该学号已存在，不能重复添加")
            else:
                name = input("请输入学生姓名：")
                age = input("请输入学生年龄：")
                sex = input("请输入学生性别：")
                score = input("请输入学生成绩：")
                grade = input("请输入学生年级：")
                data.loc[id]=[name,age,sex,score,grade]
                print("添加成功")
        except BaseException:
                print("发生异常，添加失败")  
        print("学生信息打印：", data)
        
    elif step == 2:
        id = int(input("请输入要删除的学生学号："))
        if id in data.index:
            data=data.drop(id)
            print("删除成功")
        else:
            print("学号不存在，删除失败")
        print("学生信息打印：", data)         
       
    elif step == 3:
        try:
            id = int(input("请输入修改的学生学号："))
            if id in data.index:
                name = input("请输入学生姓名：")
                age = input("请输入学生年龄：")
                sex = input("请输入学生性别：")
                score = input("请输入学生成绩：")
                grade = input("请输入学生年级：")
                data.loc[id]=[name,age,sex,score,grade]
                print("修改成功")
            else:
                print("学号不存在，修改失败")
        except BaseException:
                print("发生异常，添加失败")  
        print("学生信息打印：", data)
        
    elif step == 4:
        id = int(input("请输入要查询的学生学号："))
        if id in data.index:
            print("查询到的学生信息：")
            print(data.loc[id])
        else:
            print("学号不存在，查询失败")
        
    elif step == 5:
        key=input("请输入排序键：")
        if key=='学号':
            print(data.sort_index())
        else:
            print(data.sort_values(by=key))
    else:
        flag = 1
print("退出系统成功")
```
