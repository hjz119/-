```python
import pandas as pd
pima=pd.read_csv('pima.csv')
pima.head()
```

```python
import matplotlib.pyplot as plt
ax = pima[pima['Class variable (0 or 1)']==0].plot(x='Body mass index (weight in kg/(height in m)^2)', y='Triceps skin fold thickness (mm)', kind='scatter', c='red', ax=None, label='0')
pima[pima['Class variable (0 or 1)']==1].plot(x='Body mass index (weight in kg/(height in m)^2)', y='Triceps skin fold thickness (mm)', kind='scatter', c='green' , ax=ax, label='1')
plt.savefig('1.png')
```

```python
color = {1:'red', 0:'green'}
pd.plotting.scatter_matrix(pima.iloc[:,2:6], figsize=(15, 15), diagonal='kde', marker='o',
s=40, alpha=0.4, c=pima['Class variable (0 or 1)'].apply(lambda x: color[x]))
plt.savefig('2.png')
```

```python
from pandas.plotting import andrews_curves
andrews_curves(pima,'Class variable (0 or 1)',color=('green','red'))
plt.savefig('3.png')
```
